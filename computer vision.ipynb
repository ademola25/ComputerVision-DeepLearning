## FACE AND EYE DETECTION

```python
# load the face and eye cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
# Create a function that will be used to detect the face
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        
        
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3)
        for(ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (255,0,0), 2)
        
    return frame
```

```python
# Detecting face with our webcam
video_capture = cv2.VideoCapture(0) #0 is your laptop webcam while 1 is external webcam

while True:
    _, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow("Video", canvas)
    
    if cv2.waitKey(1) & 0xff ==ord("q"):
        break
video_capture.release
cv2.destroyAllWindows()
```

## SMILE DETECTION

```python
# load the face, eye and smile cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')
```

```python
# Create a function that will be used to detect the face
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        
        
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3)
        for(ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (0,0,255), 2)
            
        smile = smile_cascade.detectMultiScale(roi_gray, 1.7, 22)
        for(sx,sy,sw,sh) in smile:
            cv2.rectangle(roi_color, (sx,sy), (sx+sw, sy+sh), (0,255,0), 2)
        
    return frame
```

```python
# Detecting face with our webcam
video_capture = cv2.VideoCapture(0) #0 is your laptop webcam while 1 is external webcam

while True:
    _, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow("Video", canvas)
    
    if cv2.waitKey(1) & 0xff ==ord("q"):
        break
video_capture.release
cv2.destroyAllWindows()
```

## Deep LEarning - HAnd Detection

```python
#!pip3 install opencv-python
#!pip3 install mediapipe

#Import libraries
import cv2
import mediapipe as mp
```

```python
# Create a video object
cap = cv2.VideoCapture(0) # 0 is for webcam, 1 is external webcam
# Leverage the mediapipe library to identify hands
mphands = mp.solutions.hands
hands = mphands.Hands()
mpDraw = mp.solutions.drawing_utils
```

```python
while True:
    success, img = cap.read()
    imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(imgGray)
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                #print(id, lm)
                h,w,c = img.shape
                cx,cy = int(lm.x*w), int(lm.y*h)
                print(id, cx, cy)
                if id in [0,4,8,12,16,20]:
                    cv2.circle(img, (cx,cy), 15, (255,0,255), cv2.FILLED)
            mpDraw.draw_landmarks(img, handLms, mphands.HAND_CONNECTIONS)
    cv2.imshow('Video', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
print("done")
```
